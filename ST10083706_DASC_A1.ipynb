Identify Duplicates by track_id by counting duplicates of each track:

```python
count_track = df['track_id'].value_counts()

# Display tracks that appear more than once
duplicate_tracks = count_track[count_track > 1]
print(duplicate_tracks)
```

Analysis for categorical columns by identifying and displaying the sum of the unique values for each categorical column:

```python
df.columns
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns

for column in categorical_columns:
    print(f"Column: {column}")
    print(df[column].value_counts().head(10))  
    
    print('\n')
```

Analysis for numerical columns by identifying and displaying the sum of the unique values:

```python
numerical_columns = df.select_dtypes(include=[np.number]).columns

print(df[numerical_columns].describe())
```

Distribution of numerical columns:

```python
for column in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], bins=30, kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
